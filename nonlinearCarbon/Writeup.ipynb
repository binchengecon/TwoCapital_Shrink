Globally averaged temperature:

$$
c\frac{dT}{dt} = R_{in} -  R_{out}(T, C)
$$

$$
R_{in}(T) = Q_0(1 - p \alpha_\text{land} - p \alpha_\text{oc}(T)) 
$$

$$
R_{out}(T, C) = \kappa (T - T_\kappa) - a \ln (\frac{C}{C_0}) 
$$

$$
\alpha_{\mathrm{oc}}(T)=\left\{\begin{array}{ccc}
\alpha_{\max } & \text { if } & T \leqslant T_{\alpha, l}, \\
\alpha_{\max }+\frac{\alpha_{\min }-\alpha_{\max }}{T_{\alpha, u}-T_{\alpha, l}}\left(T-T_{\alpha, l}\right) & \text { if } & T_{\alpha, l}<T \leqslant T_{\alpha, u}, \\
\alpha_{\min } & \text { if } & T_{\alpha, u}<T .
\end{array}\right.
$$

Accumulated anthropogenic carbon, $G$
$$
\frac{d G}{dt} = g

$$

0 & \text { if } & T_{v, h}<T .
\end{array}\right.
$$




$$
T_{o, l}(G)=\left\{\begin{array}{ccc}
T_{o, l}^{-} & \text {if } & G \leqslant G_{l}, \\
T_{o, l}^{-}+\frac{\left(T_{o, l}^{+}-T_{o, l}^{-}\right)\left(G-G_{l}\right)}{G_{h}-G_{l}} & \text { if } & G_{l}<G \leqslant G_{h}, \\

\end{array}\right.
$$




$$
T_{v, l}(G)=\left\{\begin{array}{ccc}
T_{v, l}^{-} & \text {if } & G \leqslant G_{l}, \\

T_{v, l}^{+} & \text {if } \quad G_{h}<G .
\end{array}\right.
$$




$$
\beta(T)=\left\{\begin{array}{clc}
1 & \text { if } & T \leqslant T_{\alpha, l}, \\
